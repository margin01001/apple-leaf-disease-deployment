```python
X = pickle.load(open('/content/drive/MyDrive/Data/Leaf Disease Loaded data/X.pkl', 'rb'))
y = pickle.load(open('/content/drive/MyDrive/Data/Leaf Disease Loaded data/y.pkl', 'rb'))

X = X/255.0

label = []
print(y[:15])
for x in y:
  if x == 'Black_rot':
    label.append(0)
  elif x == 'Cedar_apple_rust':
    label.append(1)
  elif x == 'Apple_scab':
    label.append(2)
  elif x == 'healthy':
    label.append(3)
```
